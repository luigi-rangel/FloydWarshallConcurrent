```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./results/results.csv')
df.head()
```

```python
df.rename(columns={'Unnamed: 0': 'nthreads'}, inplace=True)
```

```python
for i in range(4):
    seq = df[str(i)]
    seq = seq[seq.index == 0]
    df["Aceleracao " + str(i)] = float(seq)/df[str(i)]
df
```

```python
size = int((df.shape[1] -1)/2)
```

```python
for i in range(4):
    df["Eficiencia " + str(i)] = df["Aceleracao " + str(i)]/df['nthreads']
```

```python
df
```

```python
for i in range(4):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
    
    ax1.plot( df['nthreads'], df["Aceleracao " + str(i)], marker='o', linestyle='-', color='b', label='')
    ax1.set_xlabel('Numero de threads')
    ax1.set_ylabel('Acelerao')
    ax1.set_title('Acelerao Teste ' + str(i))
    ax1.legend()

    ax2.plot(df["nthreads"], df["Eficiencia " + str(i)], marker='o', color='r', label='')
    ax2.set_xlabel('Numero de threads')
    ax2.set_ylabel('Eficincia')
    ax2.set_title('Eficincia Teste ' + str(i))
    ax2.legend()
    
    plt.tight_layout()
    #plt.show()
    plt.savefig('desempenho' + str(i))
```
